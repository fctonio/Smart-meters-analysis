# Look at half hourly meter readings for each household

```python
import pandas as pd
```

Combine multiple CSV files into one dataframe

```python
combined_csv = pd.concat([pd.read_csv(f'../Data/hhblock_dataset/block_{i}.csv') for i in range(112)])
```

```python
combined_csv
```

Merge dataframe with 'informations_households.csv'

```python
info_csv = pd.read_csv('../Data/informations_households.csv')
```

```python
merged_csv = combined_csv.merge(info_csv, on = 'LCLid')
```

```python
merged_csv
```

Look into unknown class 'ACORN-U'

```python
merged_csv.loc[merged_csv['Acorn'] == 'ACORN-U']
```

```python
merged_csv = merged_csv.groupby(by ='Acorn').mean().reset_index()
```
